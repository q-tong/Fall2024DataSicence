## Non-blocking output of kernel

If you execute the next cell, you will see the output arriving as it is generated, not all at the end.

```python
import time, sys
for i in range(8):             #print numbers from 0 to 7 with a delay of 0.5 seconds between each print. 
    print(i, end = ''),        #Numbers will be printed on the same line, separated by spaces, because of the end=' '
    time.sleep(0.5)            #After each number is printed, the code will pause for 0.5 seconds
```

### Mathematics

And we also support the display of mathematical expressions typeset in LaTeX, which is rendered
in the browser thanks to the [MathJax library](http://mathjax.org).  

Note that this is *different* from the above examples.  Above we were typing mathematical expressions
in Markdown cells (along with normal text) and letting the browser render them; now we are displaying
the output of a Python computation as a LaTeX expression wrapped by the `Math()` object so the browser
renders it.  The `Math` object will add the needed LaTeX delimiters (`$$`) if they are not provided:

```python
from IPython.display import Math
Math(r'F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx')  ## 'r'denotes a raw string in python
```

With the `Latex` class, you have to include the delimiters yourself.  This allows you to use other LaTeX modes such as `eqnarray`:

```python
from IPython.display import Latex
Latex(r"""\begin{eqnarray}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\
\nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0 
\end{eqnarray}""")
```

```python
# %load http://matplotlib.sourceforge.net/mpl_examples/pylab_examples/integral_demo.py
#!/usr/bin/env python

# implement the example graphs/integral from pyx
from pylab import *
from matplotlib.patches import Polygon

def func(x):
    return (x-3)*(x-5)*(x-7)+85

ax = subplot(111)          # create subplot(111), which means a single subplot is placed in a 1x1 grid.

a, b = 2, 9                # integral area
x = arange(0, 10, 0.01)    # x is defined as an array of values ranging from 0 to 10 with a step of 0.01. 
y = func(x)                # y is calculated by applying the function func(x) to each value in the array x.
plot(x, y, linewidth=1)    # plot of the function using the x-values (x) and their corresponding y-values (y). 
                           # The linewidth=1 parameter sets the thickness of the plotted line.


# make the shaded region
ix = arange(a, b, 0.01) 
iy = func(ix)
verts = [(a,0)] + list(zip(ix,iy)) + [(b,0)]  #starts with the point (a, 0), then includes all the points created by zip(ix, iy), which pairs each x-value in ix with its corresponding y-value in iy to create points along the curve of the function, then ends with the point (b, 0).  
poly = Polygon(verts, facecolor='0.8', edgecolor='k') #Creating Polygon use verts; facecolor='0.8' sets the color of the interior of the polygon to a light gray (on a scale where 0 is black and 1 is white); edgecolor='k' sets the color of the edges of the polygon to black.
ax.add_patch(poly)  #Adding the Polygon to the Plot:


# Add the integral text annotation
text(0.5 * (a + b), 30,
     r"$\int_a^b f(x)\mathrm{d}x$", horizontalalignment='center',
     fontsize=20)

axis([0,10, 0, 180])
figtext(0.9, 0.05, 'x')
figtext(0.1, 0.9, 'y')
ax.set_xticks((a,b))
ax.set_xticklabels(('a','b'))
ax.set_yticks([])
show()
```
